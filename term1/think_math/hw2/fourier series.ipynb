```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def partial_sum(size, x):
    n = np.arange(1, size)
    a0 = 1 / 6 
    an = -1 / (np.pi * n) ** 2
    c = np.cos(2 * np.pi * n * x)
    
    return a0 + an.dot(c)
```

```python
grid = np.arange(0, 1, 0.001)
partial_sums = [partial_sum(4, x) for x in grid]
original = [x - x ** 2 for x in grid]

plt.plot(grid, partial_sums)
plt.plot(grid, original)
```

```python
plt.plot(grid, np.abs(np.array(partial_sums) - np.array(original)))
```

```python
def partial_sum(size, x):
    n = np.arange(1, size)
    a0 = 1 / 2
    bn = -2 * np.cos(2 * np.pi * n) / (2 * np.pi * n)
    c = np.sin(2 * np.pi * n * x)
    
    return a0 + bn.dot(c)

grid = np.arange(0, 1, 0.001)
partial_sums = [partial_sum(5, x) for x in grid]
original = [x for x in grid]

plt.plot(grid, partial_sums)
plt.plot(grid, original)
```

```python
plt.plot(grid, np.abs(np.array(partial_sums) - np.array(original)))
```
